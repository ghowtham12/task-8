```python
import pandas as pd
df = pd.read_csv('/content/English.csv')
df
```

```python
df_1 = df.iloc[:80000,:]
df_2 = df.iloc[80000:,:]
df1 =df_1.head(100)
```

```python
df2= df_2.reset_index(drop=True)
df2.head()
```

```python
!pip install textblob
```

```python
from textblob import Word

result=[]
for i in df1['English words/sentences']:  
    
    word = Word(i)
    
    a = word.correct()
    
    result.append(''.join(a))


```

```python
se =pd.Series(result)
wor=pd.DataFrame(se, columns=["corrected words"])
```

```python
from textblob import TextBlob
result=[]
for i in df2['English words/sentences']: 
  sentence = TextBlob(i)
    
  b = sentence.correct()
    
  result.append(''.join(b))
```

```python
sc =pd.Series(result)
sent=pd.DataFrame(sc, columns=["corrected sentence"])
```

```python
spell_correction = pd.concat([df1,wor,df2,sent],axis=1)
spell_correction.head()
```
